```python
def divide_chunks(l, n): 
    for i in range(0, len(l), n):  
        yield l[i:i + n] 
```

```python
chunks = list(divide_chunks(new_records, 500))
i = 0
for chunk in chunks:
    with open(f"la_non_bnf_records_label_ok_{i:03d}.mrc", 'wb') as out:
        for record in chunk:
            out.write(record.as_marc())
        i += 1
```
